```python
dfs = pd.read_excel('Bases BIDS a esquentar - Copia.xlsx', sheet_name=None)
for name, df in dfs.items():
    globals()[f"tabela_{name.replace(' ', '_').lower()}"] = df
```

```python
dfs
```

```python
tabela1 = dfs['Bids WTW 19 e 20']
tabela1 = tabela1.drop(['ID'], axis=1)
```

```python
tabela1
```

```python
tabela2 = dfs['BD Geral WTW']
tabela2 = tabela2.drop(['ID'], axis=1)
```

```python
tabela3 = dfs['Contatos WTW']
tabela3 = tabela3.drop(['ID'], axis=1)
```

```python
tabela4 = dfs['Mailling Geral WTW']
tabela4 = tabela4.drop(['ID'], axis=1)
```

```python
tabela5 = dfs['CRM RJ 20']
tabela6 = dfs['CRM SP 20']
tabela7 = dfs['Conta Loc 24']
tabela8 = dfs['Contact Loc 24']
tabela9 = dfs['IT SEG']
tabela10 = dfs['ITSEG especifico']
tabela11 = dfs['Oportunidad LOC 24']
```

```python
bids = tabela1.merge(tabela2, on=['Nome da Empresa', 'Tabela'], how='outer')
```

```python
bids
```

```python
bids.columns
```

```python
bids = bids.merge(tabela3, on=['Nome da Empresa', 'Tabela'], how='outer')
bids = bids.merge(tabela4, on=['Nome da Empresa', 'Tabela'], how='outer')
bids = bids.merge(tabela5, on=['Nome da Empresa', 'Tabela'], how='outer')
bids = bids.merge(tabela6, on=['Nome da Empresa', 'Tabela'], how='outer')
bids = bids.merge(tabela7, on=['Nome da Empresa', 'Tabela'], how='outer')
bids = bids.merge(tabela8, on=['Nome da Empresa', 'Tabela'], how='outer')
bids = bids.merge(tabela9, on=['Nome da Empresa', 'Tabela'], how='outer')
bids = bids.merge(tabela10, on=['Nome da Empresa', 'Tabela'], how='outer')
bids = bids.merge(tabela11, on=['Nome da Empresa', 'Tabela'], how='outer')
```
